**Data Cleaning**
- Process of detecting faults or inconsistency in dataset by scrapping or modifying them to fit the definition of quality for analysis.

**Common types of data quality problems:**
1. Missing values
2. Duplicate data
3. Incorrect data types
4. Outliers and anomalies
5. Inconsistent formats
6. Typo errors

**Data Cleaning tasks**
1. Handling missing data
2. Removing duplicates
3. Correcting inaccuracies
4. Standardizing formats
5. Dealing with outliers

**Data Preprocessing**
- involves cleaning, transforming and organizing raw data to ensure it is accurate, consistent and ready for modeling.

**Implementation of data preprocessing**
1. Import Libraries and Load Dataset
*pd.read_csv('Geeksforgeeks/Data/diabetes.csv')*

2. Inspect Data Structure and Check Missing Values

*df.info()*: Prints concise summary including count of non-null entries and data type of each column.

*df.isnull().sum()*: Returns the number of missing values per column.

3. Statistical Summary and Visualizing Outliers

*df.describe()*: Computes count, mean, std deviation, min/max and quartiles for numerical columns.

4. Remove Outliers Using the Interquartile Range (IQR) Method

5. Correlation Analysis

*df.corr()*: Computes pairwise correlation coefficients between columns.

6. Visualize Target Variable Distribution

*plt.pie()*: Pie chart to display proportion of each class in the target variable 'Outcome'.

7. Separate Features and Target Variable

*df.drop(columns=[...])*: Drops the target column from features.

8. Feature Scaling: Normalization and Standardization

- Normalization (Min-Max Scaling): Rescales features between 0 and 1. Good for algorithms like k-NN and neural networks.

- Standardization: Transforms features to have mean = 0 and standard deviation = 1, useful for normally distributed features.

**Tokenization** is a core preprocessing technique, especially for text data, that involves breaking down a continuous stream of text into smaller units called tokens.
